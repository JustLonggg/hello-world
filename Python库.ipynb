```python
#线性回归
def residual(t, x, y):
    return y - (t[0] * x ** 2 + t[1] * x + t[2])
from scipy.optimize import leastsq
x = np.linspace(-2,2,50)
A,B,C = 2,3,-1
y = (A*x**3 + B*x**2 + C) + np.random.rand(len(x))*0.75
t = leastsq(residual,[0,0,0],args=(x,y))
theta = t[0]
y_hat = (theta[0]*x**2 + theta[1]*x + theta[2])
plt.plot(x,y,'r-',lw=2,label='Actual')
plt.plot(x,y_hat,'g-',lw=2,label='Predict')
plt.legend(loc='upper left')
plt.grid()
plt.show()
```

# SVD

```python
import os 
from PIL import Image
import matplotlib as mpl
from pprint import pprint

def restore(sigma,u,v,K): #奇异值，左特征向量、右特征向量
    m = len(u)
    n = len(v[0])
    a = np.zeros((m,n))
    for k in range(K):
        uk = u[:,k].reshape(m,1)
        vk = v[k].reshape(1,n)
        a += sigma[k] * np.dot(uk,vk)
    a[a<0] = 0
    a[a>255] = 255
    return np.rint(a).astype('uint8')

A = Image.open('6.son.png','r')
output_path = r'.\Pic'
if not os.path.exists(output_path):
    os.mkdir(output_path)
a = np.array(A)
K = 50
u_r,sigma_r,v_r = np.linalg.svd(a[:,:,0])
u_g,sigma_g,v_g = np.linalg.svd(a[:,:,1])
u_b,sigma_b,v_b = np.linalg.svd(a[:,:,2])
plt.figure(figsize=(10,10),facecolor='w')
mpl.rcParams['font.sans-serif'] = [u'simHei']
mpl.rcParams['axes.unicode_minus'] = False
for k in range(1,K+1):
    print (k)
    R = restore(sigma_r,u_r,v_r,k)
    G = restore(sigma_g,u_g,v_g,k)
    B = restore(sigma_b,u_b,v_b,k)
    I = np.stack((R,G,B),2)
    Image.fromarray(I).save('%s\\svd_%d.png' % (output_path,k))
    if k <= 12:
        plt.subplot(3,4,k)
        plt.imshow(I)
        plt.axis('off')
        plt.title(u'奇异值个数：%d' % k)
plt.suptitle(u'SVD与图像分解',fontsize=18)
plt.tight_layout(2)
plt.subplots_adjust(top=0.9)
plt.show()
```

# 卷积（convolve）

```python
#时间序列中的应用（简单移动平均和指数移动平均）
stock_max, stock_min, stock_close, stock_amount = np.loadtxt('6.SH600000.txt',delimiter='\t',skiprows=2,usecols=(2,3,4,5),unpack=True)
N = 100
stock_close = stock_close[:N]
n = 5
weight = np.ones(n)
weight /= weight.sum()
stock_sma = np.convolve(stock_close,weight,mode='valid') #简单移动平均

weight = np.linspace(1,0,n)
weight = np.exp(weight)
weight /= weight.sum()
stock_ema = np.convolve(stock_close,weight,mode='valid') #指数移动平均

mpl.rcParams['font.sans-serif'] = [u'SimHei']
mpl.rcParams['axes.unicode_minus'] = False

t = np.arange(n-1,N)
plt.plot(np.arange(N),stock_close,'r-',lw=2,label='原始收盘价')
plt.plot(t,stock_sma,'g-',lw=2,label='简单移动平均')
plt.plot(t,stock_ema,'b-',lw=2,label='指数移动平均')
plt.legend(loc='upper right')
plt.title('股票收盘价')
plt.grid(True)
plt.show()
```

```python
#卷积的图像处理
def convolve(image,weight):
    height,width = image.shape
    h,w = weight.shape
    height_new = height-h+1
    width_new = width-w+1
    image_new = np.zeros((height_new,width_new),dtype=np.float)
    for i in range(height_new):
        for j in range(width_new):
            image_new[i,j] = np.sum(image[i:i+h,j:j+w] * weight)
    image_new = image_new.clip(0,255)
    image_new = np.rint(image_new).astype('uint8')
    return image_new

A = Image.open("6.son.png", 'r')
output_path = '.\\Pic\\'
if not os.path.exists(output_path):
    os.mkdir(output_path)
a = np.array(A)
soble_x = np.array(([-1, 0, 1], [-2, 0, 2], [-1, 0, 1]))
soble_y = np.array(([-1, -2, -1], [0, 0, 0], [1, 2, 1]))
soble = np.array(([-1, -1, 0], [-1, 0, 1], [0, 1, 1]))
prewitt_x = np.array(([-1, 0, 1], [-1, 0, 1], [-1, 0, 1]))
prewitt_y = np.array(([-1, -1,-1], [0, 0, 0], [1, 1, 1]))
prewitt = np.array(([-2, -1, 0], [-1, 0, 1], [0, 1, 2]))
laplacian = np.array(([0, -1, 0], [-1, 4, -1], [0, -1, 0]))
laplacian2 = np.array(([-1, -1, -1], [-1, 8, -1], [-1, -1, -1]))

weight_list = ('soble_x', 'soble_y', 'soble', 'prewitt_x', 'prewitt_y', 'prewitt', 'laplacian', 'laplacian2')

print('梯度检测：')
for weight in weight_list:
    print(weight,'R')
    R = convolve(a[:,:,0],eval(weight))
    print('G')
    G = convolve(a[:,:,1],eval(weight))
    print('B')
    B = convolve(a[:,:,2],eval(weight))
    
    I = 255-np.stack((R,G,B),2)
    Image.fromarray(I).save(output_path + weight + '.png')
```
