 <font size=4 color=red face='黑体'>使函数能够使用于数组

```python
def bmi(h,w):
    bmi_index = w/h ** 2
    if bmi_index > 18 and bmi_index < 25:
        return 0
    elif bmi_index <= 18:
        return -1
    else:
        return 1
h = (2.2-1.4) * np.random.random_sample((10,)) + 1.4  #产生0到1的随机数（1行10列）
h = np.round(h,2)
w = (96-30)*np.random.random_sample((10,))+30
w = np.round(w,2)
print(w)
#对数组来说，逻辑运算符and不再适用
bmi_ufunc = np.frompyfunc(bmi,2,1) #2表示参数个数，1表示返回值个数
bmi_ufunc(h,w)
```

<font color=red size=4 face='黑体'>数组直接相乘：对应位置直接相乘；矩阵直接相乘：相当于数组的np.dot（标量积）；np.inner:内积；np.cross:矢量积；np.tensordot：张量积

```python
a = np.arange(1,10).reshape(3,3)
b = np.array([[1,0,1],[0,1,1],[1,1,0]])
A = np.mat(a)
B = np.mat(b)
print(a*b)
print('===========')
print(np.dot(a,b))
print('===========')
print(A*B)
print('===========')
print(np.inner(a,b))
print('===========')
print(np.cross(a,b))
print('===========')
print(np.tensordot(a,b))
```

```python
from matplotlib.ticker import MultipleLocator,FormatStrFormatter

t = np.linspace(0,100,100)
s = 9.8*t**2/2
fig,ax = plt.subplots(figsize=(8,4))
ax.plot(t,s)

ax.set_xlabel('time')
ax.set_xlim((0,100))
ax.set_ylabel('distance')

xmajor_locator = MultipleLocator(20)
xmajor_formatter = FormatStrFormatter('%1.1f')
xminor_locator = MultipleLocator(5)
ax.xaxis.set_major_locator(xmajor_locator)
ax.xaxis.set_major_formatter(xmajor_formatter)
ax.xaxis.set_minor_locator(xminor_locator)

ymajor_locator = MultipleLocator(10000)
ymajor_formatter = FormatStrFormatter('%1.1f')
yminor_locator = MultipleLocator(5000)
ax.yaxis.set_major_locator(ymajor_locator)
ax.yaxis.set_major_formatter(ymajor_formatter)
ax.yaxis.set_minor_locator(yminor_locator)

ax.grid(True,which='minor')

for tick in ax.xaxis.get_major_ticks():
    tick.label1.set_fontsize(17)
```

```python
names = ['movie_id','title','genres']
movies = pd.read_table(r'F:\pydata-book-2nd-edition\pydata-book-2nd-edition\datasets\movielens\movies.dat',
                      sep='::',header=None,names=names)
movies[:10]
```

```python
#从genres中取出所有的genre值
all_genres = []
for i in movies.genres:
    all_genres.extend(i.split('|'))
genres = pd.unique(all_genres)
genres
```

```python
#构建全0的DataFrame
dummies = pd.DataFrame(np.zeros((len(movies),len(genres))),columns=genres)
dummies.head()
```

```python
for i,gen in enumerate(movies.genres):
    indices = dummies.columns.get_indexer(gen.split('|'))
    dummies.iloc[i,indices] = 1
movies_gen = movies.join(dummies.add_prefix('Genre_'))
movies_gen.iloc[0]
```

<font size=5 color=red face='黑体'>正则表达式

```python
import re
text = 'foo     bar\t baz    \tqux'
re.split('\s+',text)
```

```python
#用re.compile编译可重用的regex对象
regex = re.compile('\s+')
regex.split(text)
```

```python
regex.findall(text)
```

```python
#findall、search、match的区别
text = '''
Dave dave@google.com
Steve steve@gmail.com
Rob rob@gmail.com
Ryan ryan@yahoo.com
'''
pattern = r'[A-Z0-9._%+-]+@[A-Z0-9.-]+\.[A-Z]{2,4}'
regex = re.compile(pattern,flags=re.IGNORECASE)   #不区分大小写
regex.findall(text)   #findall返回所有
```

```python
#search返回第一个
s = regex.search(text)
s
```

```python
text[s.start():s.end()]
```

```python
#match只匹配出现在字符串开头的模式
print(regex.match(text))
```

```python
#sub替换
print(regex.sub('REDACTED',text))
```

```python
#分组：（）
pattern = r'([A-Z0-9._%+-]+)@([A-Z0-9.-]+)\.([A-Z]{2,4})'
regex = re.compile(pattern,flags=re.IGNORECASE)
m = regex.match('wesm@bright.net')
m.groups()
```

```python
regex.match('wesm@bright.net').groups()
```

```python
regex.findall(text)
```

```python
print(regex.sub(r'Username:\1,Domain:\2,Suffix:\3',text))
```
