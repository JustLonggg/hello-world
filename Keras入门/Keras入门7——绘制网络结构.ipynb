```python
from keras.datasets import mnist
from keras.utils import np_utils
from keras.models import Sequential
from keras.layers import Dense,Dropout,Convolution2D,MaxPooling2D,Flatten
from keras.optimizers import Adam
from keras.utils.vis_utils import plot_model
import matplotlib.pyplot as plt
```

```python
model = Sequential([
    Convolution2D(input_shape=(28,28,1),filters=16,kernel_size=4,strides=1,padding='same',activation='relu',name='conv1'),
    MaxPooling2D(pool_size=2,strides=2,padding='same',name='pool1'),
    Convolution2D(filters=32,kernel_size=4,strides=1,padding='same',activation='relu',name='conv2'),
    MaxPooling2D(2,2,'same',name='pool2'),
    Flatten(),
    Dense(500,activation='relu'),
    Dropout(0.4),
    Dense(10,activation='softmax')
])
```

```python
plot_model(model,to_file='F:\keras\model.png',show_shapes=True,show_layer_names=True,
          rankdir='TB')  #TB表示从上到下，LR表示从左到右
plt.figure(figsize=(10,10))
img = plt.imread('F:\keras\model.png')
plt.imshow(img)
plt.axis('off')
plt.show()
```
